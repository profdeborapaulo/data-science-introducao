```python
#Gráfico de Barras
#Bom para representar frequências.
#Segue um array para representar as pessoas e suas frequências

x = np.array(['Julia', 'Felipe', 'Elaine','Ed Carlos']) #simulando pessoas
y = np.array([2, 8, 3, 10]) #frequÊncia delas

plt.bar(x,y)
plt.show()
```

```python
#Gráfico de Barras Horizontais

x = np.array(['Julia', 'Felipe', 'Elaine','Ed Carlos']) #simulando pessoas
y = np.array([2, 8, 3, 10]) #frequÊncia delas

plt.barh(x,y)
plt.show()
```

```python
# Gráfico de Dispersão:4#Gráfico de Dispersão
#Nuvens de pontos para duas variáveis

x = np.array([5, 6, 9, 8, 7, 4, 3 , 2, 1])
y = np.array([55, 76, 89, 88, 99, 44, 35, 24, 16])

plt.scatter(x,y)
```
